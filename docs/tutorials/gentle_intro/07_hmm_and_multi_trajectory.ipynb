# Part 7: HMMs and multiple trajectories

This notebook closes the [Gentle Introduction to Dynestyx](../00_index.ipynb) by pointing to two important extensions: **hidden Markov models (HMMs)** with discrete latent states, and **multiple trajectories** (e.g. hierarchical or panel data).

## 7.1 Hidden Markov models (HMMs)

When the **latent state is discrete** (e.g. $x_t \in \{1,\ldots,K\}$) rather than continuous, the same `DynamicalModel` interface applies: you provide an **initial condition** (e.g. `Categorical`), a **state evolution** that returns a distribution over the next state (e.g. `Categorical(A[x])`), and an **observation model**. Dynestyx provides **`FilterBasedHMMMarginalLogLikelihood`** for exact computation of the marginal log-likelihood via the forward algorithm, so you can do parameter inference (e.g. NUTS or SVI) over transition and emission parameters without sampling the discrete state sequence.

**Full tutorial:** [HMM inference](../../hmm_inference.ipynb)

## 7.2 Multiple trajectories

You may have **multiple time series** (e.g. several experiments or subjects). Then you typically share some parameters across trajectories and possibly use hierarchical priors. Fitting to multiple trajectories is supported by building a model that, for each trajectory, sets a context and calls `dsx.sample` (or uses a single context per trajectory in a loop/plate).
