# A Gentle Introduction to Dynestyx

This multi-part tutorial is a conceptual complement to the [Quickstart](../../quickstart.ipynb). We start from basic Bayesian inference in NumPyro, then show how **dynestyx** extends it to sample and infer **dynamical systems**‚Äîwith a clear separation between *what* the model is (parameters + dynamics + observations) and *how* we simulate or compute likelihoods (simulators and filters).

## Contents

1. **[Part 1: NumPyro and the Bayesian workflow](../01_numpyro_bayesian_workflow/)** ‚Äî Linear regression with NUTS and ArviZ; generating data with `Predictive`; prior and posterior predictive (Bayesian workflow).

2. **[Part 2: Dynestyx ‚Äî discrete-time dynamical systems](../02_dynestyx_discrete_intro/)** ‚Äî What dynestyx is; separation of concerns; a first model (discrete-time stochastic volatility); context and handlers; generating data with `DiscreteTimeSimulator`; NUTS + simulator for $p(x, \theta \mid \text{data})$.

3. **[Part 3: Filtering and the marginal log-likelihood](../03_filtering_mll/)** ‚Äî Computing MLL with the particle filter at $\theta_{\text{mean}}$ and $\theta_{\text{true}}$; switching to Taylor KF; canonical reference placeholder.

4. **[Part 4: Filtering + NUTS ‚Äî pseudomarginal inference](../04_filtering_nuts_pseudomarginal/)** ‚Äî Using the filter‚Äôs MLL inside NUTS to sample only parameters; when this is ‚Äúoverkill‚Äù vs. valuable (with links to be filled).

5. **[Part 5: SVI and warming up NUTS](../05_svi/)** ‚Äî SVI as a fast alternative; using SVI to initialize NUTS; execution times; rapid prototyping.

6. **[Part 6a: Continuous-time dynamical systems (SDEs)](../06_continuous_time/)** ‚Äî `ContinuousTimeStateEvolution` (drift, diffusion coefficient, diffusion covariance); `SDESimulator`; Lorenz 63 with partial observations; `LinearGaussianObservation` and $H$; NUTS + EnKF; full-observations deep dive link.

7. **[Part 6b: Continuous-time dynamical systems (ODEs)](../06b_odes/)** - `ContinuousTimeStateEvolution` (drift and no diffusions); `ODESimulator`. Probabilistic numerics coming soon!

8. **[Part 7: Hidden Markov Models (HMMs)](../07_hmm/)** ‚Äî Working with categorical state spaces models (filtering and Bayesian parameter estimation).

9. **[Part 8: Hierarchical / mixed-effect inference](../08_hierarchical_inference/)** ‚Äî `plate` for trajectory-level parameters; hierarchical dynamical models and inference.

10. **[Part 9: Discrete-time smoothing](../09_discrete_smoothing/)** ‚Äî `Smoother` for discrete-time models; smoothed state summaries and future-only posterior rollout.

11. **[Part 10: Continuous-time smoothing](../10_continuous_smoothing/)** ‚Äî Continuous-discrete smoothing; smoothed state summaries and future-only posterior rollout.

12. **[Part 11: Missing observations with Filters and Smoothers](../11_missing_observations/)** ‚Äî Whole-row and partial missingness with the cuthbert Kalman filter/smoother path.

13. **[Part 11b: Missing observations with `LatentStateBuilder` + MCMC](../11b_missing_observations_latent_state_builder_mcmc/)** ‚Äî Joint posterior inference over parameters and latent states under Gaussian missing-data latent-state-builder conditioning.

14. **[Part 11c: Missing observations in HMMs](../11c_missing_observations_hmms/)** ‚Äî Exact HMM missingness handling for `MultivariateNormal` and `Independent(..., 1)` observation families.
