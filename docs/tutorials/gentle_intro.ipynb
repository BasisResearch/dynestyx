# A Gentle Introduction to Dynestyx

This tutorial has been **split into a multi-part series** for easier reading and navigation.

**→ Start here: [Gentle Introduction — Index](gentle_intro/00_index.ipynb)**

The series covers:

1. NumPyro and the Bayesian workflow (linear regression, NUTS, ArviZ, Predictive, prior/posterior predictive)
2. Dynestyx discrete-time intro (separation of concerns, stochastic volatility, context, handlers, NUTS + simulator)
3. Filtering and the marginal log-likelihood (PF, Taylor KF, MLL at θ_mean and θ_true)
4. Filtering + NUTS (pseudomarginal inference)
5. SVI and warming up NUTS
6. Continuous-time dynamical systems (ContinuousTimeStateEvolution, SDESimulator, L63 partial obs, EnKF)
7. HMMs and multiple trajectories (with links to dedicated tutorials)
